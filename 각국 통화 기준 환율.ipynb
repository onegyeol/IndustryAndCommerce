```python
import pandas as pd

file_path = '주요국_통화의_대미달러환율.csv' 
df = pd.read_csv(file_path)

#평균자료만 추출
df_avg = df[df['측정항목'] == '평균자료'].copy()

# 월별 컬럼만 추출
date_columns = df.columns[5:]
df_avg[date_columns] = df_avg[date_columns].apply(pd.to_numeric, errors='coerce')

# "통화당 달러" -> 역수 계산 (달러당 통화로)
usd_per_currency_mask = df_avg['단위'] == '통화당 달러'
df_avg.loc[usd_per_currency_mask, date_columns] = df_avg.loc[usd_per_currency_mask, date_columns].apply(
    lambda col: col.map(lambda x: 1 / x if pd.notnull(x) and x != 0 else None)
)

def convert_account_label(label, unit):
    if '/' in label:
        a, b = label.split('/')
        if unit == '통화당 달러':
            return f"USD/{b.strip()}"
        elif unit == '달러당 통화':
            return f"USD/{a.strip()}"
    return label

labels = df_avg.apply(lambda row: convert_account_label(row['계정항목'], row['단위']), axis=1)
df_avg = df_avg.copy()
df_avg['표기'] = labels

converted_df = df_avg[['표기'] + list(date_columns)].set_index('표기')

# 전체 환율 테이블 출력
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(converted_df)
```

```python
import pandas as pd

file_path = '주요국_통화의_대미달러환율.csv' 
df = pd.read_csv(file_path)

df_avg = df[df['측정항목'] == '평균자료'].copy()

# 📅 월별 컬럼만 추출
date_columns = df.columns[5:]
df_avg[date_columns] = df_avg[date_columns].apply(pd.to_numeric, errors='coerce')

# 모든 월이 NaN인 행 제거
df_avg = df_avg.dropna(subset=date_columns, how='all').copy()

#  "통화당 달러" -> 역수 변환 (달러당 통화로)
usd_per_currency_mask = df_avg['단위'] == '통화당 달러'
df_avg.loc[usd_per_currency_mask, date_columns] = df_avg.loc[usd_per_currency_mask, date_columns].apply(
    lambda col: col.map(lambda x: 1 / x if pd.notnull(x) and x != 0 else None)
)

def convert_account_label(label, unit):
    if '/' in label:
        a, b = label.split('/')
        if unit == '통화당 달러':
            return f"USD/{b.strip()}"
        elif unit == '달러당 통화':
            return f"USD/{a.strip()}"
    return label

labels = df_avg.apply(lambda row: convert_account_label(row['계정항목'], row['단위']), axis=1)
df_avg['표기'] = labels

converted_df = df_avg[['표기'] + list(date_columns)].set_index('표기')

# 전체 출력
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(converted_df)
```
